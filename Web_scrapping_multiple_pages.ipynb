```python
# Imports
import re
import pandas as pd
```

```python
def set_pattern(text):
    '''This function changes the pattern of the page'''
    
    # Patter for matching
    pattern = re.compile(text)
    
    return pattern
```

```python
def create_df(name):
    '''This function creates DataFrame with given name as parameter and headers Nazwa and Opis'''
    
    # Create DataFrame
    headers = ['Nazwa','Opis']
    name = pd.DataFrame(index=range(0,1), columns=headers)
    
```

```python
def clean_lists():
    '''This function cleans the A and B list for further creation of DataFrame'''
    
    # Declare lists
    A = []
    B = []
```

```python
def url_open_and_use_soup(url):
    '''This function opens given url and returns the BeautifulSoup object of this page'''
    
    page = urllib2.urlopen(url)
    soup = BeautifulSoup(page)
    
    return soup
```

```python
def find_all_h1(soup):
    '''This function find all headers 1st type and add its text to the list A'''
    
    soup_headers = soup.findAll('h1')
    
    for h in soup_headers:
        A.append(h.find(text=True))
```

```python
def find_all_description(soup):
    '''This function find the description of the word and add its text to the list A'''
    
    soup_headers = soup.find_all("div", {"class" : "definicja"})

    for h in soup_headers:
        B.append(h.find(text=True))
```

```python
def check_page(soup):
    '''This function takes BeautifulSoup object and prints out all the urls found in it'''
    
    for link in soup.findAll("a"):
            url = link["href"]
            absolute_url = urljoin(base_url, url)
            check_pattern(url, absolute_url)
            print(url, absolute_url)
```

```python
def check_pattern(url, absolute_url, text = "slownik/a[^\s]*$"):
    '''print all urls with words matching pattern set as first parameter and 
    then opening the pages to implement findAll functions on BeautifulSoup object
    to find h1 and descriptions'''
    
    pattern = set_pattern(text)
    if pattern.match(url):
        #print(url, absolute_url)
        page = urllib2.urlopen(absolute_url)
        soup1 = BeautifulSoup(page)
        find_all_h1(soup1)
        find_all_description(soup1)  
```

```python
def fill_df(name = "df"):
    '''This function creates DataFrame with the given name and then fill it with the data from lists A and B and prints it'''
    create_df(name)
    print("Df name is: " + name)
    
    name = pd.DataFrame(A, columns=["Słowo"])
    name["Opis"] = B
    
    print(name)
```
