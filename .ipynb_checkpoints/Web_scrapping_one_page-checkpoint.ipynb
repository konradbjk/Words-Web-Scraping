```python
# Importing necessary libraries
import urllib.request as urllib2
from bs4 import BeautifulSoup
```

```python
url = "https://blog.winezja.pl/slownik-terminow-winiarskich/"
```

```python
page = urllib2.urlopen(url)
```

```python
soup = BeautifulSoup(page)
```

```python
print(soup.prettify())
```

```python
all_spans = soup.find_all("span", {"style" : "font-family: Tahoma;"})

for l in all_spans:
    #print(l.get("strong"))
    print(l)
```

```python
#Get all the names
all_strongs = soup.find_all("strong")
```

```python
del all_strongs[0]
for l in all_strongs:
    print(l.string)
```

```python
# To get two columns
A = []
B = []
```

```python
for r in all_spans:
    #print(r.contents[1])
    names = r.findAll('strong')
    description = r.contents[1]
    description = description[2:]
    B.append(description)
    for e in names:
        A.append(e.find(text=True))
    

print(B)
```

```python
# create DF
import pandas as pd
```

```python
df = pd.DataFrame(A, columns=["Słowo"])
df["Opis"] = B
df
```

```python
df.to_csv("slownik.csv", encoding="UTF-8")
```

## Translation

```python
import numpy as np
from googletrans import Translator
data = pd.read_csv('slownik.csv')

headers = ['A','B','A_translation', 'B_translation']
df2 = pd.DataFrame(index=range(0,len(data)), columns=headers)
```

```python
# Initialize class object
translator = Translator()
```

```python
def translate_row(row):
    ''' Translate elements A and B within `row`. '''
    a = translator.translate(row[1], dest='en')
    b = translator.translate(row[2], dest='en')
    return pd.Series([a.origin, b.origin, a.text, b.text], headers)
```

```python
for i, row in enumerate(data.values):
    # Fill empty dataframe with given serie.
    df2.loc[i] = translate_row(row)
```
